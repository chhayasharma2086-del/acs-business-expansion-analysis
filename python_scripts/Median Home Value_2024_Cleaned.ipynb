```python
#Import Liabries
import pandas as pd
```

```python
#load DataSet
df = pd.read_csv(r"C:\Users\chhay\Documents\Project\American Community Survey (ACS)\Raw_Data\Median Home Value 2024(1 Year).csv")
```

```python
#Data Overview
df.head()
df.info()
```

```python
#Remove Margin of Column
df = df.loc[: ,~df.columns.str.contains("Margin of Error")]
df.head()
```

```python
#Rename Columns name
df.rename(columns ={"Label (Grouping)": "Metric","Estimate": "Median_Home_Value"}, inplace=True)
df.head()
```

```python
#Reshape/Transpose Data
homevalue =df.melt(id_vars="Metric",
                var_name="State",
                value_name="Median_Home_Value")
homevalue.head()
```

```python
#Remove “United States” Total Row
homevalue = homevalue[homevalue['State'] != 'United States!!Estimate']
homevalue.head()
```

```python
#Clean State Name 
homevalue["State"]=homevalue["State"].str.replace("!!Estimate","")
homevalue.head()
```

```python
#Remove Commas From Numbers
homevalue['Median_Home_Value'] = (
    homevalue['Median_Home_Value']
    .replace(',', '', regex=True)
)
```

```python
#Convert Columns to Numeric
homevalue['Median_Home_Value'] = pd.to_numeric(
    homevalue['Median_Home_Value'],
    errors='coerce'
)
```

```python
#Keep Only Columns need
homevalue = homevalue[['State','Median_Home_Value']]
homevalue.head()
```

```python
#Check Missing Values
homevalue.isnull().sum()
```
