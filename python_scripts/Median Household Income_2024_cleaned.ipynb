```python
#Import Liaberies
import pandas as pd
```

```python
#Load dataSet
df = pd.read_csv(r"C:\Users\chhay\Documents\Project\American Community Survey (ACS)\Raw_Data\Median Household Income 2024(1 Year).csv")
```

```python
#Data Overview
df.head()
df.info()
```

```python
#Remove Margin of Error Columns
df = df.loc[: ,~df.columns.str.contains("Margin of Error")]
df.head()
```

```python
#Rename Columns Name
df.rename(columns ={"Label (Grouping)": "Metric", "Estimate": "Median_Income"}, inplace=True)
df.head()
```

```python
#Reshape/Transponse Data
income= df.melt(id_vars="Metric",
                  var_name="State",
                  value_name="Median_Income")
income.head()
```

```python
#Remove “United States” Total Row
income = income[income['State'] != 'United States!!Estimate']
income.head()
```

```python
#Clean State Name 
income["State"]=income["State"].str.replace("!!Estimate","")
income.head()
```

```python
#Remove Commas From Numbers
income['Median_Income'] = (
    income['Median_Income']
    .replace('[\$,]', '', regex=True)
)
```

```python
#Convert Columns to Numeric
income['Median_Income'] = pd.to_numeric(
   income['Median_Income'],
    errors='coerce'
)
```

```python
#Keep Only Columns need
income = income[['State', 'Median_Income']]
income.head()
```

```python
#Check Missing Values
income.isnull().sum()
```

```python
#Data Overview
income.head()
income.info()
```

```python
#save the cleaned Data
income.to_csv(r"C:\Users\chhay\Documents\Project\American Community Survey (ACS)\Cleaned_Data\Income_cleaned.csv", index=False)
```
