```python
# IMPORT LIBRARIES
import pandas as pd
```

```python
# LOAD DATASET
df = pd.read_csv(r"C:\Users\chhay\Documents\Project\American Community Survey (ACS)\Raw_Data\Educational Attainment 2024(1 Year).csv")
```

```python
# DATA OVERVIEW / INSPECTION
df.head()
df.info()
```

```python
# REMOVE MARGIN OF ERROR COLUMNS
df = df.loc[:, ~ df.columns.str.contains("Margin of Error")]
df.head()
```

```python
# RENAME MAIN LABEL COLUMN
df.rename(columns={"Label (Grouping)": "Education_Level"}, inplace = True)
df.head()
```

```python
# RESHAPE/Transpose DATA (WIDE → LONG FORMAT)
education = df.melt( id_vars=["Education_Level"],
                  var_name="State",
                  value_name="Estimate")
education.head()
```

```python
#Check Unique Education Levels
education['Education_Level'].unique()
```

```python
#Remove Weird Characters
education['Education_Level'] = (education['Education_Level'].str.replace('재', '', regex=False).str.strip()
)
education.head()
```

```python
#Remove United States Total
education = education[
    education['State'] != 'United States!!Estimate'
]
education.head()
```

```python
# CLEAN STATE NAMES
education["State"] = education["State"].str.replace("!!Estimate", "")
education.head()
```

```python
#Keep Only Important Degrees
important_levels = [
    'Regular high school diploma',
    "Bachelor's degree",
    "Master's degree",
    'Doctorate degree'
]

education = education[
    education['Education_Level'].isin(important_levels)
]
education.head()
```

```python
#Rename Education Categories
education['Education_Level'] = education[
    'Education_Level'
].replace({
    'Regular high school diploma': 'High_School',
    "Bachelor's degree": 'Bachelors',
    "Master's degree": 'Masters',
    'Doctorate degree': 'Doctorate'
})
education.head()
```

```python
#Remove Commas From Numbers
education['Estimate']= (
    education['Estimate'].replace('[\$,]', '', regex=True)
)
```

```python
#Convert Columns to Numeric
education['Estimate'] = pd.to_numeric(
    education['Estimate'],
    errors='coerce'
)
```

```python
#Pivot Table
education_cleaned = education.pivot_table(
    index='State',
    columns='Education_Level',
    values='Estimate',
    aggfunc='sum'
).reset_index()
education_cleaned.head()
```
