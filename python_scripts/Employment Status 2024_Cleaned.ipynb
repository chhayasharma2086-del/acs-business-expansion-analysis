```python
# Import Liabraries
import pandas as pd
```

```python
# Load Dataset
df = pd.read_csv(r"C:\Users\chhay\Documents\Project\American Community Survey (ACS)\Raw_Data\Employment Status 2024(1 Year).csv")
```

```python
# Data Overview/ Inspection
df.head()
df.info()
```

```python
# Remove Margin of error column
df = df.loc[: , ~df.columns.str.contains("Margin of Error")]
df.head()
```

```python
# Rename Label(Grouping) column
df.rename(columns={"Label (Grouping)": "Employment_Status"},inplace= True)
df.head()
```

```python
# reshape/Transpose Data
employment = df.melt(
    id_vars="Employment_Status",
    var_name="State",
    value_name="Estimate")
employment.head()
```

```python
#Check Unique Education Levels
employment["Employment_Status"].unique()
```

```python
#Remove Weird Characters
employment["Employment_Status"] = (employment["Employment_Status"].str.replace('재', '', regex=False).str.strip())
employment.head()
```

```python
# Clean State Name 
employment["State"]= employment["State"].str.replace("!!Estimate", "")
employment.head()
```

```python
#Remove United States Total
employment = employment[
    (employment['State'] != 'United States') &
    (employment['Employment_Status'] != 'Total:')
]
employment.head()
```

```python
#Pivot Table
employment = employment.pivot_table(
    index='State',
    columns='Employment_Status',
    values='Estimate',
    aggfunc='sum'
).reset_index()
employment.head()
```

```python
#Rename Columns
employment = employment.rename(columns={
    'Civilian labor force:': 'Labor_Force',
    'In labor force:': 'Total_Labor_Force',
    'Not in labor force': 'Not_In_Labor_Force'
})
employment.head()
```

```python
#Remove Commas From Numbers
cols = employment.columns.drop('State')

employment[cols] = employment[cols].replace(',', '', regex=True)

employment.head()
```

```python
#Convert Columns to Numeric
employment[cols] = employment[cols].apply(
    pd.to_numeric,
    errors='coerce'
)
employment.head()
```

```python
employment.info()
```

```python
#Create Employment Rate
employment['Employment_Rate'] = (
    employment['Employed'] /
    employment['Labor_Force']
)
```

```python
#Create Unemployment Rate
employment['Unemployment_Rate'] = (
    employment['Unemployed'] /
    employment['Labor_Force']
)
```

```python
employment['Employment_Rate'] = (
    employment['Employment_Rate'] * 100
).round(2)

employment['Unemployment_Rate'] = (
    employment['Unemployment_Rate'] * 100
).round(2)
```

```python
#Keep only useful columns
employment_cleaned = employment[[
    'State',
    'Labor_Force',
    'Employed',
    'Unemployed',
    'Employment_Rate',
    'Unemployment_Rate'
]]
```
