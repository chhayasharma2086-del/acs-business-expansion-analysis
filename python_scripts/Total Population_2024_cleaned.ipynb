```python
#Import Liabaries
import pandas as pd
```

```python
#Load Dataset
df = pd.read_csv(r"C:\Users\chhay\Documents\Project\American Community Survey (ACS)\Raw_Data\Total Population 2024(1 Year).csv")
```

```python
#Data Overview
df.head()
df.info()
```

```python
#Remove Margin of Error Column
df = df.loc[:,~df.columns.str.contains("Margin of Error")]
df.head()
```

```python
#Rename Columns name
df.rename(columns = {"Label (Grouping)":"Category", "Estimate":"Population"},inplace = True)
df.head()
```

```python
#Reshape/Transpose of Data
population = df.melt(id_vars="Category",
                  var_name="State",
                  value_name="Population")
population.head()
```

```python
#Remove “United States” Total Row
population = population[population["State"] != 'United States!!Estimate']
population.head()
```

```python
#Clean State Name 
population["State"]=population["State"].str.replace("!!Estimate","")
population.head()
```

```python
#Keep Only Columns need
population = population[['State','Population']]
population.head()
```

```python
#Remove Commas
population['Population'] = (
    population['Population']
    .replace(',', '', regex=True)
)
```

```python
#Convert to Numeric
population['Population'] = pd.to_numeric(
    population['Population'],
    errors='coerce'
)
```

```python
population.head()
population.info()
```

```python
#Check Missing Values
population.isnull().sum()
```

```python
#save the cleaned Data
population.to_csv(r"C:\Users\chhay\Documents\Project\American Community Survey (ACS)\Cleaned_Data\Population_cleaned.csv", index=False)
```
